# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Smoothing data function

```python
def __smooth_data(df_):
    df_aux = df_[['val_acc','test_acc','pred_acc']]
    df_aux = df_aux.rolling(window=100).mean()
    return df_aux


def get_best_arch(df):
    df_aux = df.sort_values(by='val_acc', ascending=False)
    display(df_aux.head(10))
    best_arch_index, best_arch_val_acc = df_aux.loc[0,'id']-1, df_aux.loc[0,'val_acc']
    return best_arch_index, best_arch_val_acc


def display_plot(orig_df):
    best_arch_idx,best_arch_val_acc = get_best_arch(orig_df)
    
    df = orig_df.sort_values(by='id')
    new_df = __smooth_data(df)
    
    plt.plot(best_arch_idx, best_arch_val_acc, 'o', label="best_arch")
    plt.plot(new_df.val_acc.values , color='b')
    plt.plot(new_df.pred_acc.values, color='r')
    plt.plot(new_df.test_acc.values, color='green')
    plt.legend(['best_arch','val_acc','pred_acc','test_acc'], loc='center right')
    plt.show()
```

# Cifar-10

## 50h
