```python
import pandas as pd
import numpy as np
```

```python
algos = ['rl','random','evolutino']
mths = [1,10,50,100,200]
datasets = ['cifar10','cifar100','ImageNet16-120']
ss_indicators = ['sss','tss']

for ss in ss_indicators:
    for a in algos:
        for mth in mths:
            for ds in datasets:
                df = pd.read_csv(f'./data/{ss}/{a}_{mth}h_{ds}.csv')
                mean_train_acc = round(np.mean(df.train_accuracy),4)
                mean_val_acc = round(np.mean(df.val_acc),4)
                mean_test_acc = round(np.mean(df.test_acc),4)
                print(80*'+')
                print(f'Algo: {a} | Dataset: {ds} | Mth: {mth} | SS: {ss} | Mean Training Accuracy: {mean_train_acc}')
                print(f'Algo: {a} | Dataset: {ds} | Mth: {mth} | SS: {ss} | Mean Validation Accuracy: {mean_val_acc}')
                print(f'Algo: {a} | Dataset: {ds} | Mth: {mth} | SS: {ss} | Mean Test Accuracy: {mean_test_acc}')
                print(80*'-')
```
