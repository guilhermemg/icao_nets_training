# Import libraries

```python
import os
import pandas as pd
```

# Read labels dataframe

```python
labels_df = pd.read_csv('/home/guilherme/tensorflow_datasets/download/manual/celeb_a/2.0.1/list_attr_celeba.txt', sep=' ')
labels_df.head()
```

```python
labels_df = labels_df.drop(columns=['Unnamed: 41'])
```

```python
labels_df.head()
```

# Read partitions/split dataframe

```python
part_path = '/home/guilherme/tensorflow_datasets/download/manual/celeb_a/2.0.1/list_eval_partition.txt'
part_df = pd.read_csv(part_path, sep=' ')
part_df.head()
```

# Merge part_df and labels_df

```python
final_df = pd.merge(labels_df, part_df, on='img_name')
print(final_df.shape)
final_df.head()
```

```python
def get_split_name(sp_val):
    if sp_val == 0:
        return 'train'
    elif sp_val == 1:
        return 'validation'
    elif sp_val == 2:
        return 'test'

final_df['split'] = final_df['split'].apply(lambda x: get_split_name(x))
final_df.head()
```

```python
final_df.groupby('split').count()
```

```python
cols = [x for x in final_df.columns if not x in ['img_name','split']]
final_df = final_df[['img_name','split']+cols]
final_df.head()
```

# Replace label value -1 to 0

```python
final_df = final_df.replace(-1, 0)
final_df.head()
```

# Create train_df, devel_df, test_df

```python
train_df = final_df[final_df.split == 'train']
devel_df = final_df[final_df.split == 'validation']
test_df = final_df[final_df.split == 'test']
```

# Save labels dataframe

```python
dest_path_train_labels = '/home/guilherme/data1/Dropbox/Link to Desktop/Doutorado/Datasets/celeb_a/train_data.csv'
dest_path_devel_labels = '/home/guilherme/data1/Dropbox/Link to Desktop/Doutorado/Datasets/celeb_a/valid_data.csv'
dest_path_test_labels = '/home/guilherme/data1/Dropbox/Link to Desktop/Doutorado/Datasets/celeb_a/test_data.csv'

train_df.to_csv(dest_path_train_labels, index=False)
devel_df.to_csv(dest_path_devel_labels, index=False)
test_df.to_csv(dest_path_test_labels, index=False)
```
